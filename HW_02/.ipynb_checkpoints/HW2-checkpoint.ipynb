## Задание №1. Горячая точка (алгоритм - 10 баллов, визуализация - 10 баллов).

Генерируйте рандомные точки на планете Земля до тех пор, пока не попадете на территорию ``Афганистана``

1. Вы можете использовать функции принадлжености точки полигону и расстояния от точки до полигона (в метрах)
2. Предложите не наивный алгоритм поиска (генерировать __напрямую__ точку из полигона границ Афганистана __запрещено__)

```python
import folium
from shapely.geometry import Point
import random
from OSMPythonTools.nominatim import Nominatim
from shapely import wkt


#граница Афганистана в формате geometry Json
nominatim = Nominatim(cacheDir='cache')
geometryJson = nominatim.query('Afghanistan', wkt=True).toJSON()


map = folium.Map(tiles='Stamen Terrain', location=([33.7680065, 66.2385139]), zoom_start=3)


#граница Земли
minLat = -90
maxLat = 90
minLon = -180
maxLon = 180


#рандомная точка из рамки
p = Point(random.uniform(minLat, maxLat), random.uniform(minLon, maxLon))
folium.Marker([p.x, p.y], popup='<i>Poin 1</i>', tooltip=f'{p.x} {p.y}</i>').add_to(map)


#рамка границы Афганистана
boundsPolygon = wkt.loads(geometryJson[0]['geotext'])
minBlon, minBlat, maxBlon, maxBlat = boundsPolygon.bounds


#пока точка не в границах Афганистана, идет сужение
i = 1; 
while boundsPolygon.contains(Point(p.y, p.x)) == False:
    i += 1
    if p.x > maxBlat:
        maxLat = p.x
    elif p.x < minBlat:
        minLat = p.x
        
    if p.y > maxBlon:
        maxLon = p.y
    elif p.y < minBlon:
        minLon = p.y
        
    p = Point(random.uniform(minLat, maxLat), random.uniform(minLon, maxLon))
    folium.Marker([p.x, p.y], popup=f'<i>Poin {i}</i>', tooltip=f'{p.x} {p.y}</i>').add_to(map)
    
# обычно меньше чем за 20 бросков попадаем в Афганистан
folium.GeoJson(boundsPolygon).add_to(map)


map
```

## Задание №3. Поездка по Нью-Йорку (маршрут - 20 баллов, визуализация - 10 баллов).

Добраться __на автомобиле__ от входа в ``Central Park`` __Нью-Йорка__ (со стороны ``5th Avenue``) до пересечения ``Water Street`` и ``Washington Street`` в Бруклине (откуда получаются лучшие фото Манхэттенского моста) довольно непросто - разумеется, из-за вечных пробок. Однако еще сложнее это сделать, проезжая мимо школ, где дети то и дело переходят дорогу в неположенном месте.

Вам необходимо построить описанный выше маршрут, избегая на своем пути школы. Визуализируйте данный маршрут (также добавив школы и недоступные для проезда участки дорог) при помощи ``Folium``

Данные о расположении школ Нью-Йорка можно найти [здесь](https://catalog.data.gov/dataset/2019-2020-school-point-locations)

```python
import folium
from openrouteservice import client
from shapely.geometry import LineString, mapping #, Point
from OSMPythonTools.nominatim import Nominatim
from OSMPythonTools.overpass import overpassQueryBuilder, Overpass
from shapely.ops import cascaded_union 


def style_function(color): # To style data
    return lambda feature: dict(color=color, opacity=0.5, weight=4,)


map = folium.Map(location=([40.7815, -73.9429]), zoom_start=11)
api_key = '5b3ce3597851110001cf6248bf801aa32b0b4124b48ff0dcef7ecc63'
clnt = client.Client(key=api_key)


popup_route = "<h4>{0} route</h4><hr>" \
             "<strong>Duration: </strong>{1:.1f} mins<br>" \
             "<strong>Distance: </strong>{2:.3f} km" 


# Путь
coordinates = [[-73.94973,40.79618], [-73.98959,40.70320]]


# Построение пути без учета школ "зеленый путь"
direction_params = {'coordinates': coordinates,
                    'profile': 'driving-car', 
                    'format_out': 'geojson',
                    'preference': 'shortest',
                    'geometry': 'true'}
old_route = clnt.directions(**direction_params)

duration, distance = old_route['features'][0]['properties']['summary'].values()
popup = folium.map.Popup(popup_route.format('Old Route', duration/60, distance/1000))

folium.GeoJson(old_route, name='Old Route', style_function=style_function('green')).add_child(popup).add_to(map)
folium.Marker(list(reversed(coordinates[0])), popup='Input Central Park').add_to(map)
folium.Marker(list(reversed(coordinates[1])), popup='Water Street and Washington Street').add_to(map)


# Поиск школ и увеличение границ полигона
schools = []

nominatim = Nominatim(cacheDir='cache')
areaId = nominatim.query('Manhattan Island, New York').areaId()

overpass = Overpass(cacheDir='cache')
query = overpassQueryBuilder(area=areaId, elementType=['way', 'relation'], selector='"building"="school"', includeGeometry=True)
result = overpass.query(query, timeout=60)
for key in result.elements():
    if len(key.geometry()['coordinates'][0]) > 2:
        school = LineString(key.geometry()['coordinates'][0]).buffer(0.0004)
        #school = Point(school.centroid.coords[0]).buffer(0.0009)
        schools.append(school)
        folium.GeoJson(school).add_to(map)

            
buffer = cascaded_union(schools)


# Построение пути c учетом школ "черный путь"
direction_params = {'coordinates': coordinates,
                    'format_out': 'geojson',
                    'preference': 'shortest',
                    'geometry': 'true',
                    'options': {'avoid_polygons': mapping(buffer)}}
new_route = clnt.directions(**direction_params)

distance, duration = new_route['features'][0]['properties']['summary'].values()
popup = folium.map.Popup(popup_route.format('New Route', duration/60, distance/1000))

folium.GeoJson(new_route, name='New Route', style_function=style_function('black')).add_child(popup).add_to(map)


map
```
